```python
# Import libraries:
import numpy as np
from matplotlib import pyplot as plt

# Load data from txt file:
file = 'Yuri-David-File/G4bl-Output/profile.txt'
data = np.loadtxt(file)
```

```python
# Values along channel:
Z = []
emitX = []
emitY = []
emitZ = []
for i in range(len(data)):
    Z.append(data[i][0])
    emitX.append(data[i][6])
    emitY.append(data[i][7])
    emitZ.append(data[i][8])
```

```python
# Plot emittance along z:
plt.figure(figsize=(12,5))
plt.scatter(Z,emitX,color='red',label='x',s=0.15)
plt.scatter(Z,emitY,color='blue',label='y',s=0.15)
plt.scatter(Z,emitZ,color='green',label='z',s=0.15)
plt.xlabel('z (mm)')
plt.ylabel('emittance (mm-rad)')
plt.legend()
plt.show()
```

```python
# Plot zoomed-in region:
start = 0
stop = start + 100
plt.figure()
plt.plot(Z[start:stop],emitX[start:stop],color='red',label='x',marker='.')
plt.plot(Z[start:stop],emitY[start:stop],color='blue',label='y',marker='.')
plt.xlabel('z (mm)')
plt.ylabel('emittance (mm-rad)')
plt.legend()
plt.show()
```
