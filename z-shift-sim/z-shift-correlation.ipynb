```python
from matplotlib import pyplot as plt
```

```python
z_shift = [0, 125, 250, 375, 500] #cm
pi_shift = [2.25, 1.75, 1.25, 0.75, 0.25]
pf_shift = [1.125, 0.625, 0.25, 2.75, 2.375]
```

```python
plt.plot(z_shift,pi_shift,color='blue',label='initial',marker='.')
plt.plot(z_shift,pf_shift,color='red',label='final',marker='.')
plt.xlabel('initial beam z offset (cm)')
plt.ylabel('p total offset')
```
