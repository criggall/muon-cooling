```python
import numpy as np
from matplotlib import pyplot as plt
import sys
import os
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(parent_dir)
from functions.read_ecalc9_data import readECALC9
from functions.read_ecalc9_data import readECALC9F
from functions.read_ecalc9_data import readEMITCALC
```

```python
# Plot solenoids?
do_sol_plot = False

# Number of solenoids:
n = 18

# Distance between solenoids:
d = 0.7 # m

# Solenoid length:
L = 0.3 # m

# Define function to plot lines indicating solenoid placement:
def plotSolenoids(ax):
    for i in range(n):
        center = 0.5+i*d
        if np.mod(i,2) == 0:
            ax.axvspan(xmin=center-L/2, xmax=center+L/2, color='gray', alpha=0.1)
        else:
            ax.axvspan(xmin=center-L/2, xmax=center+L/2, color='gray', alpha=0.2)
```

### ECALC9

```python
# Read in data from ECALC9:
data9 = readECALC9('ecalc9.out')
```

```python
# Plot Bz vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['Bz'][0], marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$B_z$ (T)')
plt.show()
```

```python
# Plot beta vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['beta'][0], color='darkblue', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel(r'$\beta$ (m)')
plt.show()
```

```python
# Plot longitudinal beta vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['betaL'][0], color='darkblue', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel(r'$\beta_L$ (m)')
plt.show()
```

```python
# Plot emittances vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['eperp'][0], label='$\epsilon_{\perp}$', color='blue', marker='.', markersize=5)
plt.plot(data9['z'][0], data9['elong'][0], label='$\epsilon_L$', color='green', marker='.', markersize=5)
plt.plot(data9['z'][0], data9['e6D'][0], label='$\epsilon_{6D}$', color='black', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon$ (m)')
plt.legend()
plt.show()
```

```python
# Plot e_long vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['elong'][0], color='green', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon_L$ (m)')
plt.show()
```

```python
# Plot e_perp vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['eperp'][0], color='blue', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon_{\perp}$ (m)')
plt.show()
```

```python
# Plot 6D emittance vs. z:
fig, ax = plt.subplots(figsize=(9,5))
if do_sol_plot == True:
    plotSolenoids(ax)
plt.plot(data9['z'][0], data9['e6D'][0], color='black', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon_{6D}$ (m$^3$)')
plt.show()
```

```python
# Calculate transverse emittance and expected 6D emittance:
trans_emit = []
emit_prod = []
for i in range(len(data9['eperp'])):
    eperp = data9['eperp'][i]; elong=data9['elong'][i]
    trans_emit.append(eperp**2)
    emit_prod.append(eperp**2*elong)

# Plot 6D emittance:
plt.figure(figsize=(9,5))
plt.plot(data9['z'][0], emit_prod[0], label='$\epsilon_{\perp}^2\epsilon_{L}$', marker='.', markersize=5)
plt.plot(data9['z'][0], data9['e6D'][0], label='$\epsilon_{6D}$', marker='.', markersize=5)
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon$ (m$^3$)')
plt.legend()
plt.show()
```
