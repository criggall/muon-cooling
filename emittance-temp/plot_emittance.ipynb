```python
from compute_emittance import calculateEmittance
import numpy as np

from matplotlib import pyplot as plt

import sys
import os
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(parent_dir)
from functions.read_g4bl_data import readDetData

# Number of detector output files:
ndet = 30

# Define RF frequency:
f_rf = 325e6

# Specify version of HFOFO lattice:
version = 'frozen'

# Define reference momentum (MeV/c):
if version == 'frozen':
    p0 = 221.8
elif version == 'latest':
    p0 = 200

# Define peak on-axis B field (T):
if version == 'frozen':
    Bz0 = 3.86057
elif version == 'latest':
    Bz0 = 3.28247
```

```python
# Import z values:
z_vals = []
for i in range(ndet):
    file_path = os.path.join(parent_dir, 'hfofo-frozen/g4bl-output', f'out{i+1}.txt')
    data = readDetData(file_path)
    z = data['z'].values[0]
    z_vals.append(z)
```

```python
# Compute emittances:
rms_emittances = []; emittances = []
normalized_rms_emittances = []; normalized_emittances = []
for i in range(ndet):
    file_path = os.path.join(parent_dir, 'hfofo-frozen/g4bl-output', f'out{i+1}.txt')
    rms_emittance, emittance, normalized_rms_emittance, normalized_emittance = calculateEmittance(file_path=file_path, f_rf=f_rf, p0=p0, Bz0=Bz0)
    rms_emittances.append(rms_emittance)
    emittances.append(emittance)
    normalized_rms_emittances.append(normalized_rms_emittance)
    normalized_emittances.append(normalized_emittance)
```

### Emittances from RMS covariance matrix

```python
# Plot RMS emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[0] for i in rms_emittances], color='tab:red', marker='.')
plt.plot(z_vals, [i[1] for i in rms_emittances], color='tab:green', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```

```python
# Plot RMS emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[2] for i in rms_emittances], color='tab:blue', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```

```python
# Plot normalized RMS emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[1] for i in normalized_rms_emittances], color='tab:green', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```
