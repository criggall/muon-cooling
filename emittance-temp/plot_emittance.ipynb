```python
from compute_emittance import calculateEmittance

from matplotlib import pyplot as plt

# Number of detector output files:
ndet = 30
```

```python
# Compute emittances:
rms_emittances = []; emittances = []
normalized_rms_emittances = []; normalized_emittances = []
for i in range(ndet):
    rms_emittance, emittance, normalized_rms_emittance, normalized_emittance = calculateEmittance(f'det{i}.txt')
    rms_emittances.append(rms_emittance)
    emittances.append(emittance)
    normalized_rms_emittances.append(normalized_rms_emittance)
    normalized_emittances.append(normalized_emittance)
```

```python
import sys
import os
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(parent_dir)

from functions.read_g4bl_data import readDetData
```

```python
# Import z values:
z_vals = []
for i in range(ndet):
    data = readDetData(f'det{i}.txt')
    z = data['z'].values[0]
    z_vals.append(z)
```

### RMS emittances

```python
# Plot RMS emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[0] for i in rms_emittances], color='tab:red', marker='.')
plt.plot(z_vals, [i[1] for i in rms_emittances], color='tab:green', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```

```python
# Plot RMS emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[2] for i in rms_emittances], color='tab:blue', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```

### Emittances

```python
# Plot emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[0] for i in emittances], color='tab:red', marker='.')
plt.plot(z_vals, [i[1] for i in emittances], color='tab:green', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```

```python
# Plot emittances vs. z:
plt.figure(figsize=(9,5))
plt.plot(z_vals, [i[2] for i in emittances], color='tab:blue', marker='.')
plt.xlabel('$z$ (mm)')
plt.ylabel(r'$\epsilon$ (mm)')
plt.show()
```
