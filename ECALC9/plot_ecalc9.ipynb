```python
from matplotlib import pyplot as plt

# Import functions:
from read_ecalc9 import readECALC9

# Read in data from ECALC9:
data = readECALC9('ecalc9.out')
```

```python
# Number of solenoids:
n = 18

# Distance between solenoids:
d = 0.7 # m

# Solenoid length:
L = 0.3 # m

# Define function to plot lines indicating solenoid placement:
def plot_solenoids(ax):
    for i in range(n):
        center = 0.5+i*d
        ax.axvspan(xmin=center-L/2, xmax=center+L/2, color='gray', alpha=0.2)
```

```python
# Plot Bz vs. z:
fig, ax = plt.subplots(figsize=(12,5))
# plot_solenoids(ax)
# plt.scatter(data['z'], data['Bz'], s=5)
plt.plot(data['z'][0], data['Bz'][0])
plt.xlabel('$z$ (m)')
plt.ylabel('$B_z$ (T)')
plt.show()
```

```python
# Plot beta vs. z:
fig, ax = plt.subplots(figsize=(12,5))
# plot_solenoids(ax)
plt.scatter(data['z'], data['beta'], s=5, color='darkblue')
plt.plot(data['z'][0], data['beta'][0], color='darkblue')
plt.xlabel('$z$ (m)')
plt.ylabel(r'$\beta$ (m)')
plt.show()
```

```python
# Plot emittances vs. z:
plt.figure(figsize=(12,5))
plt.scatter(data['z'], data['eperp'], s=5, label='$\epsilon_{\perp}$',color='blue')
plt.scatter(data['z'], data['elong'], s=5, label='$\epsilon_L$',color='green')
plt.scatter(data['z'], data['e6D'], s=5, label='$\epsilon_{6D}$',color='black')
plt.xlabel('$z$ (m)')
plt.ylabel('$\epsilon$ (m)')
plt.legend()
plt.show()
```
